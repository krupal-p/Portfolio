```python
from pyspark.ml.feature import StringIndexer
from pyspark.ml.feature import VectorAssembler
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
from pyspark.sql.types import IntegerType
import re

#The following modeling is trying to explore and explain whether if there's any relation between the playtype and the outcome (playResult) and the feature importance of the playtype variables for difining and predicting the outcomes. A decision tree regressor is chosen to be the algorithum used for the modeling. 
#The playtype column contains three different variables which are play_type_sack, play_type_pass and play_type_unknown. 
#Those three different variables stand for three different strategies or outcomes of dropping back.
#A sack occurs when the quarterback is tackled behind the line of scrimmage before he can throw a forward pass, when the quarterback is tackled behind the line of scrimmage in the "pocket" and his intent is unclear, or when a passer runs out of bounds behind the line of scrimmage due to defensive pressure.
#Dropping back to pass is a passing style employed in American football. The quarterback initially takes a three-step drop, backpedaling into the pocket to make a pass.
#In other situations where when it's hard to differentiate the outcomes or the game has been interuppted due to different reasons, it will be categorized into play_type_unknown.  
#Based on personal experience, the dropping back strategies are important because they play important roles in deciding the tempo and the progress of the game especially during the earlier stages. Theoretically, the development during the early games are ususually the defining factor of the trend of the play style through the whole game. 
#During the vectoring assembling process, the other input chosen is the offenceFormation which stands for the formation used by possession team. It is chosen due to it's curcial role in deciding the play style during and right before the scoring time point.  
```

```python
#intended input feature playType and offenceFormation; and output label playResult

from pyspark.sql import functions as F
# File location and type
file_location = "/FileStore/tables/plays.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```python
  from pyspark.sql.types import *

  # Change column type
  df_1 = df.withColumn("playResult", df["playResult"].cast(IntegerType()))
  df_new = df_1.withColumn("defendersInTheBox", df_1["defendersInTheBox"].cast(DoubleType()))
  df_new.printSchema()
  df_new.select("playResult").show()
```

```python
def replace_null_in_offense_formation(offenseFormation):
  if (offenseFormation == "null"):
    return "EMPTY"
  if (not offenseFormation):
    return "EMPTY"
  return offenseFormation
 
replace_null_offForm_udf = udf(lambda offenseFormation: replace_null_in_offense_formation(offenseFormation), StringType())
 
df_new = df_new.withColumn("offenseFormation", replace_null_offForm_udf(df_new.offenseFormation))
df_new = df_new.fillna("I", "passResult")
display(df_new)
```

```python
nfl_plays_1 = StringIndexer(inputCol="offenseFormation", outputCol="offenseFormationIndex").fit(df_new).transform(df_new)
nfl_plays = StringIndexer(inputCol="playType", outputCol="playTypeIndex").fit(nfl_plays_1).transform(nfl_plays_1)
display(nfl_plays)
```

```python
from pyspark.ml.feature import VectorAssembler
vectorAssembler = VectorAssembler(inputCols = ["offenseFormationIndex","playTypeIndex"], outputCol = 'features')
nfl_plays_final = vectorAssembler.transform(nfl_plays)
nfl_plays_final1 = nfl_plays_final.select(['features', 'playResult'])
nfl_plays_final1.show(3)

```

```python
training, test = nfl_plays_final1.randomSplit([0.7, 0.3])
```

```python
from pyspark.ml.regression import DecisionTreeRegressor
dt = DecisionTreeRegressor(featuresCol ='features', labelCol = 'playResult')
dt_model = dt.fit(training)
dt_predictions = dt_model.transform(test)
dt_evaluator = RegressionEvaluator(
    labelCol="playResult", predictionCol="prediction", metricName="rmse")
rmse = dt_evaluator.evaluate(dt_predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)

#This model has a rmse of 10.0165 which is not perfect but accetable considering the random nature of the sport.
```
