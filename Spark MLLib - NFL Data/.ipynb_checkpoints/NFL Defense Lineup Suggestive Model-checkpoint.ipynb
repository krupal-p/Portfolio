```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.mllib.evaluation import RegressionMetrics
from pyspark.ml.regression import LinearRegression
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml import Pipeline
```

```python
%sql
SELECT
 personnelD
FROM nfl_plays_cleaned
group by personnelD
order by count(*) desc
limit 10
```

```python
plays_raw = spark.sql("""
select
case when passResult in ('C','I') then 0 else 1 end as passResult
, num_offense_runningbacks
, num_offense_tightends
, num_offense_widereceivers
, num_def_defline
, num_def_linebackers
, num_def_defbacks
, absoluteYardlineNumber
, personnelD
, dense_rank() over (order by personnelD) as personnelD_num
 from nfl_plays_cleaned where personnelD in (SELECT
 personnelD
FROM nfl_plays_cleaned
group by personnelD
order by count(*) desc
limit 10)
and passResult in ('IN','S')
and absoluteYardlineNumber is not null""")
display(plays_raw)
```

```python
from pyspark.ml.feature import StringIndexer
```

```python
feature_cols = [
    'num_offense_runningbacks', 'num_offense_tightends',
    'num_offense_widereceivers', 'absoluteYardlineNumber'
]
target_col = 'personnelD_num'

assembler = (
    VectorAssembler().setInputCols(feature_cols).setOutputCol("features"))

plays = assembler.transform(plays_raw)
plays_training, plays_test = plays.randomSplit([0.7, 0.3])
plays_training.cache()
plays_test.cache();
```

```python
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.evaluation import RegressionEvaluator

plays_reg_model = (
    RandomForestRegressor().setLabelCol(target_col).setFeaturesCol("features"))

paramGrid = (ParamGridBuilder().addGrid(plays_reg_model.maxDepth,
                                        [3, 5]).addGrid(
                                            plays_reg_model.numTrees,
                                            [20, 40]).build())

# What other stages exist?
stages = [plays_reg_model]

pipeline = Pipeline().setStages(stages)

cv = (CrossValidator().setEstimator(pipeline).setEstimatorParamMaps(
    paramGrid).setEvaluator(RegressionEvaluator().setLabelCol(target_col)))

plays_fitted_model = cv.fit(plays_training)
```

```python
print("The Best Parameters:\n--------------------")
print(plays_fitted_model.bestModel.stages[0])
plays_fitted_model.bestModel.stages[0].extractParamMap()
```

```python
plays_fitted_model.bestModel
```

```python
holdout = plays_fitted_model.transform(plays_test).selectExpr(
    "double(prediction) as prediction",
    "double(personnelD_num) as personnelD_num")
rm = RegressionMetrics(
    holdout.select("prediction",
                   "personnelD_num").rdd.map(lambda x: (x[0], x[1])))

print("MSE: ", rm.meanSquaredError)
print("MAE: ", rm.meanAbsoluteError)
print("RMSE Squared: ", rm.rootMeanSquaredError)
print("R Squared: ", rm.r2)
print("Explained Variance: ", rm.explainedVariance, "\n")
```

```python
holdout = (plays_fitted_model.bestModel.transform(plays_training).selectExpr(
    "prediction as raw_prediction", "double(round(prediction)) as prediction",
    "personnelD_num", """CASE double(round(prediction)) = personnelD_num 
  WHEN true then 1
  ELSE 0
END as equal"""))

display(holdout)
```
