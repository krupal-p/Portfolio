```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# dow30 = pd.read_csv('dow30.csv', header=None, squeeze=True)
dow30 = pd.read_csv('dow30.csv', header=0, index_col=0, parse_dates=True)
dow30[(dow30.index >= '2017-01-01') & (dow30.index <= '2020-01-01')]
```

```python
corr = dow30.corr()
corr.unstack().sort_values().tail(50)
corr.unstack().sort_values().head(20)
```

```python
fig, ax = plt.subplots(figsize=(11, 11))
ax = sns.heatmap(corr,
                 vmin=-1,
                 vmax=1,
                 center=0,
                 cmap=sns.diverging_palette(20, 220, n=200),
                 square=True)
ax.set_xticklabels(ax.get_xticklabels(),
                   rotation=45,
                   horizontalalignment='right')
```

```python
dow_pct_change = dow30.pct_change()
dow_std = dow_pct_change.std()
dow_std.argmax()  # most volatile stock
dow_std.max()
dow_std.argmin()  # least volatile stock
dow_std.min()
dow_pct_change['AAPL'].dropna().plot(label='AAPL')
dow_pct_change['PG'].dropna().plot()
```

```python
plt.plot(dow_pct_change['AAPL'].dropna(), label='AAPL')
plt.plot(
    dow_pct_change['PG'][dow_pct_change['PG'].index > '1980-12-15'].dropna(),
    label='PG')
plt.legend()
```
